## Proses Cleaning Data

Pada tahapan ini, dilakukan perubahan seluruh nama fitur. Hal ini dilakukan supaya data yang digunakan lebih rapih, dan juga lebih mudah untuk diolah nantinya

```python
df.columns= ['tahun',
            'demam_berdarah',
            'hepatitis',
            'malaria',
            'tuberkulosis',
            'aids',
            'hiv',
            'campak',
            'tetanus',
            'diare',
            'penduduk_muda',
            'penduduk_dewasa',
            'penduduk_lansia',
            'sampah_ditangani',
            'produksi_sampah',
            'kasus_narkoba',
            'kasus_seksual',
            'krb',
            'curah_hujan']
df
```

```python
df.info()
```

Pada proses ini data yang tidak memiliki value atau "NaN" akan diisi valuenya dengan teknik tertentu, bisa dengan cara interpolasi, ffill, bfill, dan lainnya.

```python
df['demam_berdarah'] = df['demam_berdarah'].interpolate(method='linear')
df['hepatitis'] = df['hepatitis'].interpolate(method='linear')
df['campak'] = df['campak'].interpolate(method='linear')
df['tetanus'] = df['tetanus'].interpolate(method='linear')
df
```

```python
df['tetanus'] = df['tetanus'].bfill()
df['penduduk_muda'] = df['penduduk_muda'].ffill()
df['penduduk_dewasa'] = df['penduduk_dewasa'].ffill()
df['penduduk_lansia'] = df['penduduk_lansia'].ffill()
df
```

```python
df['curah_hujan'] = df['curah_hujan'].fillna(df['curah_hujan'].median())
df
```

Fitur "krb" atau "Jumlah Kawasan Rawan Banjir" di hapus, karena data tersebut tidak lengkap serta ke validannya masih dipertanyakan, selain itu data tersebut tidak menghasilkan pengaruh yang signifikan terhadap data lainnya.

```python
df = df.drop(columns=['krb'])
df
```

## Visualisasi Data

Merupakan cara untuk menggambarkan data secara visual sehingga pola, tren, dan hubungan dalam data dapat lebih mudah dipahami dan diinterpretasikan. Tujuan utama dari visualisasi data adalah untuk menyampaikan informasi dengan cara yang jelas, efektif, dan menarik, sehingga memungkinkan pengguna untuk membuat pemahaman yang lebih baik tentang data dan mengambil keputusan yang didasarkan pada bukti.

## Grafik Pertumbuhan Penduduk Berdasarkan Range Usia

Dapat dilihat bahwa usia dominan yang menempati DIY Yogyakarta berada pada kalangan dewasa atau pada range umur 15 - 64 tahun, lalu diikuti peringkat kedua yaitu usia muda atau pada range umur 0 - 14 tahun. 

```python
plt.figure(figsize=(10, 6))
plt.bar(df['tahun'], df['penduduk_muda'], color='blue', label='0-14 Tahun')
plt.bar(df['tahun'], df['penduduk_dewasa'], color='green', bottom=df['penduduk_muda'], label='15-64 Tahun')
plt.bar(df['tahun'], df['penduduk_lansia'], color='red', bottom=df['penduduk_muda'] + df['penduduk_dewasa'], label='65+ Tahun')
plt.title('Perubahan Umur dari Tahun ke Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Penduduk')
plt.legend()
plt.grid(True)
plt.xticks(df['tahun'])
plt.show()
```

## Heatmap

Merupakan grafik yang digunakan untuk menyoroti pola dan hubungan dalam data tabular yang besar atau kompleks. Mereka membantu pengguna untuk dengan cepat mengidentifikasi pola-pola penting, korelasi, atau anomali dalam data.

```python
df = df.drop(columns=['tahun', 'sampah_ditangani', 'produksi_sampah'])
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi antar Fitur')
plt.show()
```
